```python
import pandas as pd
import numpy as np

import re
import string

import nltk

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("../dataset/processed/eda_dataset.csv")

df.head()
```

```python
lemmatizer = WordNetLemmatizer()

stop_words = set(stopwords.words("english"))
```

```python
df["clean_message"] = df["message"].str.lower()
```

```python
def remove_html(text):
    return re.sub(r"<.*?>", "", text)

df["clean_message"] = df["clean_message"].apply(remove_html)
```

```python
def remove_urls(text):
    return re.sub(r"http\S+|www\S+", "", text)

df["clean_message"] = df["clean_message"].apply(remove_urls)
```

```python
def remove_emails(text):
    return re.sub(r"\S+@\S+", "", text)

df["clean_message"] = df["clean_message"].apply(remove_emails)
```

```python
def remove_numbers(text):
    return re.sub(r"\d+", "", text)

df["clean_message"] = df["clean_message"].apply(remove_numbers)
```

```python
def remove_punctuation(text):
    return text.translate(
        str.maketrans("", "", string.punctuation)
    )

df["clean_message"] = df["clean_message"].apply(remove_punctuation)
```

```python
import nltk

nltk.download("punkt")
nltk.download("punkt_tab")
```

```python
def tokenize(text):
    return nltk.word_tokenize(text)

df["tokens"] = df["clean_message"].apply(tokenize)
```

```python
def remove_stopwords(tokens):
    return [
        word
        for word in tokens
        if word not in stop_words
    ]

df["tokens"] = df["tokens"].apply(remove_stopwords)
```

```python
def lemmatize(tokens):
    return [
        lemmatizer.lemmatize(word)
        for word in tokens
    ]

df["tokens"] = df["tokens"].apply(lemmatize)
```

```python
df["clean_message"] = df["tokens"].apply(
    lambda x: " ".join(x)
)
```

```python
df["clean_message"] = (
    df["clean_message"]
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
)
```

```python
df.to_csv(
    "../dataset/processed/preprocessed_dataset.csv",
    index=False
)

print("Preprocessed dataset saved successfully.")
```
