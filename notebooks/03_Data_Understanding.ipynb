```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("../dataset/processed/base_dataset.csv")

df.head()
```

```python
print("="*50)
print("Dataset Shape")
print("="*50)

print(f"Rows    : {df.shape[0]}")
print(f"Columns : {df.shape[1]}")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
duplicates = df.duplicated().sum()

print("Duplicate Records :", duplicates)
```

```python
df.drop_duplicates(inplace=True)

df.reset_index(drop=True, inplace=True)

print("New Shape :", df.shape)
```

```python
df["label"].value_counts()
```

```python
distribution = (
    df["label"]
    .value_counts(normalize=True)
    * 100
)

distribution.round(2)
```

```python
df["message_length"] = df["message"].apply(len)

df.head()
```

```python
df["message_length"].describe()
```

```python
longest = df["message_length"].max()

print("Longest Message Length :", longest)
```

```python
shortest = df["message_length"].min()

print("Shortest Message Length :", shortest)
```

```python
average = df["message_length"].mean()

print("Average Message Length :", round(average,2))
```
